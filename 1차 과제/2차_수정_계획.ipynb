## 2. 데이터 개선 방향

모델 성능 향상을 위해 데이터 측면에서도 보완 작업을 진행한다.

### 2.1 날씨 데이터 추가

방문자 수는 외부 환경의 영향을 크게 받을 수 있으므로, 예측 성능 향상을 위해 **날씨 데이터**를 새롭게 추가한다.  
특히 기온, 강수량, 날씨 상태 등은 서울대공원 방문 여부에 직접적인 영향을 줄 가능성이 높기 때문에 중요한 변수로 활용할 수 있다.

### 2.2 결손치 처리

기존 데이터에는 일부 기간의 결손이 존재하므로, 각 구간의 특성에 맞추어 다음과 같이 처리한다.

- **2019년 9월 7일 하루 전체 데이터 부재**
  - 해당 일자는 **연평균 및 일평균 값**을 활용하여 보정한다.

- **2017년 1월 ~ 3월 데이터 부재**
  - 단순 평균 대체보다 더 정교한 보완이 필요하므로, **CatBoost 모델을 활용한 예측 방식**으로 결손값을 채운다.

- **2016년 12월 18일 ~ 12월 31일 데이터 부재**
  - 이 구간 역시 **CatBoost 모델을 활용한 예측 방식**으로 결손값을 보완한다.

### 2.3 정답 데이터 비율 확대

학습에 사용할 수 있는 **정답 데이터의 비율을 점진적으로 늘려가며**, 데이터 양의 증가가 모델 성능에 어떤 영향을 미치는지 함께 확인할 예정이다.  
이를 통해 데이터 확보량과 예측 정확도 사이의 관계도 분석할 수 있다.

## 3. 성능지표 개선 방향

기존에는 다음과 같은 지표를 중심으로 모델 성능을 평가하였다.

- **MAE**
- **RMSE**
- **MAPE**
- **R²**

이 지표들은 예측값과 실제값의 차이를 수치적으로 확인하는 데 유용하며, 방문자 수처럼 연속적인 값을 예측하는 문제에서 기본적인 평가 기준으로 사용할 수 있다.

다만 이번 과제는 단순히 숫자를 맞추는 것만이 아니라, 실제 사용자 관점에서 **한산한 날을 잘 추천하는지**도 중요하다.  
서울대공원 방문자는 자주 방문하는 사람보다 특정 날을 골라 방문하는 경우가 많기 때문에, 사용자가 “한산할 것”이라고 기대하고 방문한 날에 실제로도 한산해야 서비스의 실질적인 가치가 높아진다.

이러한 이유로 기존 회귀 지표 외에 **Precision@K**를 추가로 활용할 예정이다.

### Precision@K

**Precision@K**는 모델이 가장 한산하다고 추천한 상위 K일 중, 실제로도 한산한 날의 비율을 의미한다.

즉, 이 지표를 통해 다음을 평가할 수 있다.

- 모델이 추천한 “한산한 날”이 실제로도 한산한가
- 사용자 입장에서 추천 결과가 얼마나 신뢰할 만한가
- 단순 오차 감소뿐 아니라 실제 활용성까지 확보되었는가

## 4. 종합 정리

2차 개선은 크게 다음 세 방향으로 진행된다.

1. **CatBoost 모델 중심의 성능 최적화**
   - 가장 우수한 성능을 보인 CatBoost를 선정
   - 하이퍼파라미터 및 모델 설정값 조정을 통해 추가 개선

2. **데이터 품질 향상**
   - 날씨 데이터 추가
   - 결손 구간에 대한 평균값 보정 및 CatBoost 기반 결손치 예측
   - 정답 데이터 비율 확대 실험

3. **평가 지표 확장**
   - 기존 회귀 지표(MAE, RMSE, MAPE, R²) 유지
   - 실제 추천 품질을 평가하기 위한 Precision@K 추가

이를 통해 단순 예측 정확도 향상뿐 아니라, 실제 사용자가 서울대공원을 방문할 날짜를 선택할 때 도움이 되는 방향으로 모델을 개선하는 것을 목표로 한다.
